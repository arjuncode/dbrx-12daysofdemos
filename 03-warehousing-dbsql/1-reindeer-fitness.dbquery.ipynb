```sql
-- Reindeer Fitness Report: Finding Our Endurance Champions
SELECT 
    r.reindeer_name,
    ROUND(AVG(t.heart_rate), 1) as avg_heart_rate,
    ROUND(AVG(t.speed_mph), 1) as avg_speed,
    COUNT(*) as flight_hours,
    -- Calculate fitness score: higher speed with lower heart rate = better fitness
    ROUND(AVG(t.speed_mph) / AVG(t.heart_rate) * 100, 1) as fitness_score,
    CASE 
        WHEN AVG(t.speed_mph) / AVG(t.heart_rate) * 100 > 65 THEN 'ðŸ† Elite Athlete'
        WHEN AVG(t.speed_mph) / AVG(t.heart_rate) * 100 > 55 THEN 'ðŸ’ª Flight Ready'
        WHEN AVG(t.speed_mph) / AVG(t.heart_rate) * 100 > 45 THEN 'ðŸŽ¯ Good Shape'
        ELSE 'ðŸ“ˆ Building Fitness'
    END as fitness_status
FROM fact_telemetry t
JOIN dim_reindeer r ON t.reindeer_id = r.reindeer_id
WHERE t.flight_status = 'FLYING'
GROUP BY r.reindeer_name
ORDER BY fitness_score DESC;
```
